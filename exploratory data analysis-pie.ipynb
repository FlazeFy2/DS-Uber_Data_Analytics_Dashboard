```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ncr_ride_bookings.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Booking Status','Vehicle Type','Payment Method','Driver Cancellation Reason']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Comparison Of {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Booking Distribution by Rate Category
df = pd.read_csv("ncr_ride_bookings.csv")

# Convert Rating to Rating Category
def categorize_rating(rate):
    if not pd.isna(rate):
        if rate >= 4:
            return "Perfect"
        elif rate >= 3:
            return "Normal"
        else:
            return "Low"
    else:
        return "Not Rated"

df["Rate Category"] = df["Customer Rating"].apply(categorize_rating)

# Count the number of booking in each category
stay_counts = df["Rate Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
stay_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Booking Distribution by Rate Category")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Booking Distribution by Time Category
df = pd.read_csv("ncr_ride_bookings.csv")

# Convert Time to datetime
df["Hour"] = pd.to_datetime(df["Time"], format="%H:%M:%S").dt.hour

# Convert Time to Time Category
def categorize_time(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

# Count the number of booking
df["Time Category"] = df["Hour"].apply(categorize_time)

time_counts = df["Time Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
time_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Booking Distribution by Time Category")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Booking Distribution by Rush Hour vs Normal Hour
df = pd.read_csv("ncr_ride_bookings.csv")

# Convert Time to datetime
df["Hour"] = pd.to_datetime(df["Time"], format="%H:%M:%S").dt.hour

# Convert Hour to Time Category
def categorize_time(hour):
    if (6 <= hour < 9) or (16 <= hour < 21):
        return "Rush Hour"
    else:
        return "Normal Hour"

df["Time Category"] = df["Hour"].apply(categorize_time)

# Count the number of booking
time_counts = df["Time Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
time_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Booking Distribution: Rush Hour vs Normal Hour")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Booking Distribution : Is Cashless?
df = pd.read_csv("ncr_ride_bookings.csv")


# Categorize Payment
def categorize_payment(payment):
    if payment == "Cash":
        return "Yes"
    else:
        return "No"

df["Is Cashless"] = df["Payment Method"].apply(categorize_payment)

# Count the number of booking
time_counts = df["Is Cashless"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
time_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Booking Distribution : Is Cashless?")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Booking Distribution by Lunch Time vs Commute Time
df = pd.read_csv("ncr_ride_bookings.csv")

# Convert Time to datetime
df["Hour"] = pd.to_datetime(df["Time"], format="%H:%M:%S").dt.hour

# Convert Hour to Activity Category (Lunch, Go to Work, Go Home, Other)
def categorize_activity(hour):
    if 6 <= hour <= 9:
        return "Morning Commute (Go to Work/School)"
    elif 11 <= hour <= 14:
        return "Lunch Time"
    elif 17 <= hour <= 20:
        return "Evening Commute (Go Home)"
    else:
        return "Other Hours"

df["Activity Category"] = df["Hour"].apply(categorize_activity)

# Count the number of booking
activity_counts = df["Activity Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
activity_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Booking Distribution: Lunch Time vs Commute Times")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Booking Distribution: Weekday vs Weekend
df = pd.read_csv("ncr_ride_bookings.csv")

# Convert Date to datetime
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d")

# Convert Date to Weekday Category
def categorize_day(date):
    if date.weekday() < 5:   # 0=Monday, 4=Friday
        return "Weekday"
    else:                   # 5=Saturday, 6=Sunday
        return "Weekend"

df["Day Category"] = df["Date"].apply(categorize_day)

# Count the number of booking
day_counts = df["Day Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
day_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Booking Distribution: Weekday vs Weekend")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
